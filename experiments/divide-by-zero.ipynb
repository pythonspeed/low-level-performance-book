```python
import os
os.environ["NUMBA_DUMP_OPTIMIZED"] = "1"
import numpy as np
from numba import njit

DATA_INT = np.array([1, 2, 3], dtype=np.int64)
DATA_FLOAT = np.array([1, 2, 3], dtype=np.float64)
```

```python
@njit
def divide(n, d):
    print("AAA")
    result = n / d
    print("BBBB")
    return result
```

```python
divide(DATA_FLOAT, np.float64(0.0))
```

```python
divide(DATA_INT, np.int64(0))
```

```python
divide(np.float64(12), np.float64(0))
```

```python
np.float64(12) / np.float64(0)
```

```python
np.float64(0) / np.float64(0)
```

```python
@njit(error_model="numpy")
def ono(x, y):
    if x // y > 0:
        return True
    else:
        return False
```

```python
print(ono(np.int64(3), np.int64(0)))
print(ono.py_func(np.int64(3), np.int64(0)))
```
